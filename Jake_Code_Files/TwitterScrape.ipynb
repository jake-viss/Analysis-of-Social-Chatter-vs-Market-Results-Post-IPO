## Gathering OTLY Twitter Mention Data
In this section we will be using SNScrape to gather twitter data surrounding OTLY's IPO date.

```python
# Import libraries.
import os
import pandas as pd
```

```python
# Using OS library and SNScrape to gather twitter mention data on OTLY, Esports Technologies, ABNB, LegalZoom, and FXLV
os.system('snscrape --jsonl --max-results 100000 --since 2021-05-20 twitter-search \"OTLY until:2021-08-20\"> text-query-otly.json')
os.system("snscrape --jsonl --max-results 100000 --since 2021-04-15 twitter-search \"Esports Technologies until:2021-07-15\" > text-query-ebet.json")
os.system("snscrape --jsonl --max-results 100000 --since 2020-12-10 twitter-search \"ABNB until:2021-03-10\" > text-query-abnb.json")
os.system("snscrape --jsonl --max-results 100000 --since 2021-06-30 twitter-search \"LegalZoom until:2021-09-30\" > text-query-lz.json")
os.system("snscrape --jsonl --max-results 100000 --since 2021-07-15 twitter-search \"FXLV until:2021-10-15\" > text-query-fxlv.json")
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
otly_tweets_df = pd.read_json('text-query-otly.json', lines=True)
ebet_tweets_df = pd.read_json('text-query-ebet.json', lines=True)
abnb_tweets_df = pd.read_json('text-query-abnb.json', lines=True)
lz_tweets_df = pd.read_json('text-query-lz.json', lines=True)
fxlv_tweets_df = pd.read_json('text-query-fxlv.json', lines=True)

# Displays first 5 entries from dataframe
display(otly_tweets_df.describe())
display(ebet_tweets_df.describe())
display(abnb_tweets_df.describe())
display(lz_tweets_df.describe())
display(fxlv_tweets_df.describe())

# Export a raw twitter data file to CSV for example.
otly_tweets_df.to_csv("Twitter Data/otly_all_tweets.csv")
```

```python
# Reduce dataframe to just include date.
tweetsbyday_otly = otly_tweets_df["date"]
tweetsbyday_ebet = ebet_tweets_df["date"]
tweetsbyday_abnb = abnb_tweets_df["date"]
tweetsbyday_lz = lz_tweets_df["date"]
tweetsbyday_fxlv = fxlv_tweets_df["date"]


# Drop time form date
tweetsbyday_otly = pd.to_datetime(tweetsbyday_otly).dt.date
tweetsbyday_ebet = pd.to_datetime(tweetsbyday_ebet).dt.date
tweetsbyday_abnb = pd.to_datetime(tweetsbyday_abnb).dt.date
tweetsbyday_lz = pd.to_datetime(tweetsbyday_lz).dt.date
tweetsbyday_fxlv = pd.to_datetime(tweetsbyday_fxlv).dt.date

# preview a couple of the DF's to make sure they are formatted correctly
display(tweetsbyday_ebet.head())
display(tweetsbyday_lz.head())
```

```python
# Export twitter data to CSV.
tweetsbyday_otly.to_csv("Twitter Data/tweetsbyday_otly.csv")
tweetsbyday_ebet.to_csv("Twitter Data/tweetsbyday_ebet.csv")
tweetsbyday_abnb.to_csv("Twitter Data/tweetsbyday_abnb.csv")
tweetsbyday_lz.to_csv("Twitter Data/tweetsbyday_lz.csv")
tweetsbyday_fxlv.to_csv("Twitter Data/tweetsbyday_fxlv.csv")
```
