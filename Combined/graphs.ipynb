```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
import os
import matplotlib.pyplot as plt
```

```python
CPNG_df = pd.read_csv(Path("CPNGStockData.csv"))
OTLY_df = pd.read_csv(Path("otly_dataframe.csv"))
BMBL_df = pd.read_csv(Path("bmbl_dateframe.csv"))

```

```python
CPNG1 = CPNG_df["count_pct_chg"].hvplot.line()
CPNG2 = CPNG_df["close_pct_chg"].hvplot.line()
CPNG3 = CPNG_df["vol_pct_chg"].hvplot.line()
CPNGplot = CPNG1 * CPNG2 * CPNG3
CPNGplot
```

```python
OTLY1 = OTLY_df["tweet_pct_chg"].hvplot.line()
OTLY2 = OTLY_df["close_pct_chg"].hvplot.line()
OTLY3 = OTLY_df["volume_pct_chg"].hvplot.line()
OTLYplot = OTLY1 * OTLY2 * OTLY3
OTLYplot
```

```python
BMBL1 = BMBL_df["count_pct_chg"].hvplot.line()
BMBL2 = BMBL_df["close_pct_chg"].hvplot.line()
BMBL3 = BMBL_df["vol_pct_chg"].hvplot.line()
BMBLplot = BMBL1 * BMBL2 * BMBL3
BMBLplot
```

```python
plt.rcParams["figure.figsize"] = [15, 8]
plt.rcParams["figure.autolayout"] = True
color1="blue"
color2="orange"

ax = CPNG_df["TweetCount"].plot(kind="bar")
ax.set_ylabel("Tweet Count", color=color1)
ax.tick_params(axis="y", labelcolor=color1)
ax2 = ax.twinx()

ax2.plot(ax.get_xticks(),
         CPNG_df["Volume"].values,
         linestyle="-",
        color=color2)
ax2.set_ylabel("Volume", color=color2)
ax2.tick_params(axis="y", labelcolor=color2)
ax2.set_title("CPNG Tweet Count & Volume by Day")

plt.show()
```
