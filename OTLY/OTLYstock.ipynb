## Gather OTLY stock data

```python
# Import required libraries and dependencies
import pandas as pd
import requests
import os
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
```

```python
# Load dotenv
load_dotenv()
```

```python
# Set API keys
ALPACA_API_KEY = os.getenv("ALPACA_API_KEY")
ALPACA_SECRET_KEY = os.getenv("ALPACA_SECRET_KEY")

# Make sure keys have imported correctly.
display(type(ALPACA_API_KEY))
type(ALPACA_SECRET_KEY)
```

```python
# Create the Alpaca API object
alpaca = tradeapi.REST(
    ALPACA_API_KEY,
    ALPACA_SECRET_KEY,
    api_version="v2")
```

```python
# Set start and end dates.
start_date = pd.Timestamp("2021-05-20", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-06-20", tz="America/New_York").isoformat()
```

```python
# Define ticker, time frame, and row limit.
ticker = "OTLY"
timeframe = "1D"
```

```python
# Get OTLY closing prices for the month after IPO.
otly_df = alpaca.get_barset(
    ticker,
    timeframe,
    start = start_date,
    end = end_date
).df

# Display sample data
otly_df
```

```python
# Export OTLY dataframe as a CSV
otly_df.to_csv('otly_stock.csv', sep=',', index=False)
```
